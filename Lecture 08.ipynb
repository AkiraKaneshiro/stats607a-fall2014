```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

Next we use ipython magic command `%matplotlib inline` to produce all plots inline in a notebook.

```python
%matplotlib inline
```

```python
x_vals = np.arange(1, 11)
```

```python
y_vals = x_vals ** 2
```

```python
plt.plot(x_vals, y_vals)
```

```python
plt.plot(x_vals, y_vals, '-r')  # red line, default is blue line
```

```python
plt.plot(x_vals, y_vals, 'or')  # red circles
```

```python
plt.plot(x_vals, y_vals, '-or')  # red line with circles
```

For more information on line styles and format strings, go [here](http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.plot).

```python
x_vals = np.arange(0, 8, .5)
plt.xlabel("My x axis label")  # note how we set the axes labels
plt.ylabel("My y axis label")
plt.plot(x_vals, x_vals, '-ob', x_vals, x_vals**2, ':^r', x_vals, x_vals**2.5, '--sg')
```
